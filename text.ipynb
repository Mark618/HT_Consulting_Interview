```python
import pandas as pd
import numpy as np
import preprocessor as p
```

```python
text_df = pd.read_csv("csv/tweets.csv")
```

```python
text_df=text_df.sort_values('likes',ascending=False).head(1000)
```

```python
text_df.shape
```

```python
def clean_text(text):
    clean_text=p.clean(text)
    return clean_text
```

```python
text_df['clean']=text_df['tweets'].apply(clean_text)
```

```python
text_dt = list(text_df['clean'].to_numpy())
print(text_dt[0:3])
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
text=" ".join(text_dt)
```

```python
wordcloud = WordCloud(width = 800, height = 800, background_color ='white', min_font_size = 10).generate(text)
```

```python
plt.figure(figsize = (8, 8), facecolor = None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad = 0)
  
plt.show()
```
