Preprocess

Drop Nan columns

```python
pre_df=merged_df.dropna(subset=['tweets','time'])
```

```python
pre_df.isnull().sum()
```

```python
pre_df.info()
```

Specify Datetime format

```python
pre_df['user_created_time'] = pd.to_datetime(pre_df['user_created_time'],format='%a %b %d %H:%M:%S +0000 %Y')
```

```python
pre_df['time']
```

```python
pre_df['time'] = pd.to_datetime(pre_df['time'],format='%Y-%m-%d %H:%M:%S%z')
```

```python
pre_df.info()
```

Feature Extraction

```python
pre_df['created_year']=pre_df['user_created_time'].dt.year
pre_df['created_month']=pre_df['user_created_time'].dt.month
pre_df['created_date']=pre_df['user_created_time'].dt.day
pre_df['created_hour']=pre_df['user_created_time'].dt.hour
```

```python
pre_df['tweet_year']=pre_df['time'].dt.year
pre_df['tweet_month']=pre_df['time'].dt.month
pre_df['tweet_date']=pre_df['time'].dt.day
pre_df['tweet_hour']=pre_df['time'].dt.hour
```

Years user has been using Twitter

```python
uniq_id=pre_df.drop_duplicates(subset=['id'],keep='last')
```

```python
day_used=pd.DataFrame({'days':uniq_id['time'].dt.date-uniq_id['user_created_time'].dt.date})
```

```python
day_used['days']=pd.to_timedelta(day_used['days'])
```

```python
day_used['days']=day_used['days'].dt.days
```

```python
day_used['years']=day_used['days']/365.25
```

```python
dy_used_df=day_used.groupby(['days','years']).size().reset_index()
```

```python
fig = px.scatter(dy_used_df, x='days', y=dy_used_df.iloc[:,2], title='Account Age in Days')
fig.update_xaxes(title='Account Age(Days)')
fig.update_yaxes(title='Number of People')
fig.update_layout(plot_bgcolor='rgba(0,0,0,0)')
fig.show()
```

Sentiment Analysis

Found out tweets contain emoji and slang that are used by Malaysian. Therefor will be using VaderSentiment which understanding of emoji.

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

analyzer = SentimentIntensityAnalyzer()
```

```python
may_df['tweets'].iloc[0]
```

```python
text_df=pre_df.drop(['id','user_created_time','time','user_profile_location','tweet_geo_loc'],axis='columns')
```

```python
sentmnt_df=pd.DataFrame(text_df['tweets'])
```

```python
def get_sentiment_scores(tweet):
    scores = analyzer.polarity_scores(tweet)
    return scores
```

```python
sentmnt_df=sentmnt_df.head(500)
```

```python
sentmnt_df['sentiment']=sentmnt_df['tweets'].apply(get_sentiment_scores)   
```

```python
sentmnt_df.head()
```

Get the mention from tweets

```python
mentioned_re =r'(?<=@)\w+'
```

```python
def mention_tweet(tweet):
    mention= re.findall(mentioned_re,tweet)
    return mention
```

```python
text_df['mention']=text_df['tweets'].apply(mention_tweet)
```
